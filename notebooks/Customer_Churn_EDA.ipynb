Business Context:


AlphaCom, a leading telecommunications provider, has recently experienced a concerning rise in customer churn despite offering competitive services and a wide product portfolio. This increase is directly impacting revenue and undermining brand reputation in an intensely competitive market. Traditional retention strategies have proven inadequate because customer churn is influenced by a complex mix of factors, including service usage, billing preferences, contract types, and demographics. Without clear insights into these patterns, the company is left reacting to churn instead of preventing it.


Business Significance:


Industry research consistently shows that the cost of acquiring a new customer is 
5–6
 times higher than retaining an existing one. Therefore, even a marginal improvement in customer retention can lead to substantial cost savings. In addition, the probability of successfully selling new products or services to existing customers is significantly higher (
60–70%
) compared to new customers (
5–20%
). This makes customer retention not only a defensive strategy, but also a strong revenue growth opportunity.


From a financial perspective, increasing customer retention by just 5% can boost profits by 
25% to 95%
, depending on the industry. At a macro level, U.S. companies are estimated to lose approximately 
$136.8
 billion annually due to avoidable customer churn, highlighting the scale of the problem. These benchmarks emphasize why understanding, predicting, and preventing churn is a high‑priority business objective.


Customer churn is not merely the loss of customers; it directly affects revenue, customer lifetime value, operational efficiency, and brand reputation. High churn rates often indicate underlying issues such as poor service quality, pricing dissatisfaction, or weak customer engagement. Hence, a data‑driven churn analysis enables organizations to move from reactive churn management to proactive retention strategies.


Objective:


As a data scientist at AlphaCom, you are tasked with developing a predictive model to identify customers at high risk of churn and uncover the key factors driving their decisions. Solving this problem will enable the company to proactively design targeted retention strategies, reduce churn-related losses, and improve customer lifetime value, ultimately safeguarding revenue and strengthening AlphaCom’s competitive position.

```python
def histogram_boxplot(data, feature, figsize=(10, 7), kde=False, bins=None):
    """
    Boxplot and histogram combined

    data: dataframe
    feature: dataframe column
    figsize: size of figure (default (15,10))
    kde: whether to show the density curve (default False)
    bins: number of bins for histogram (default None)
    """
    f2, (ax_box2, ax_hist2) = plt.subplots(
        nrows=2,  # Number of rows of the subplot grid= 2
        sharex=True,  # x-axis will be shared among all subplots
        gridspec_kw={"height_ratios": (0.25, 0.75)},
        figsize=figsize,
    )  # creating the 2 subplots
    sns.boxplot(
        data=data, x=feature, ax=ax_box2, showmeans=True, color="violet"
    )  # boxplot will be created and a triangle will indicate the mean value of the column
    sns.histplot(
        data=data, x=feature, kde=kde, ax=ax_hist2, bins=bins
    ) if bins else sns.histplot(
        data=data, x=feature, kde=kde, ax=ax_hist2
    )  # For histogram
    ax_hist2.axvline(
        data[feature].mean(), color="green", linestyle="--"
    )  # Add mean to the histogram
    ax_hist2.axvline(
        data[feature].median(), color="black", linestyle="-"
    )  # Add median to the histogram
```

```python
# function to create labeled barplots


def labeled_barplot(data, feature, perc=False, n=None):
    """
    Barplot with percentage at the top

    data: dataframe
    feature: dataframe column
    perc: whether to display percentages instead of count (default is False)
    n: displays the top n category levels (default is None, i.e., display all levels)
    """

    total = len(data[feature])  # length of the column
    count = data[feature].nunique()
    if n is None:
        plt.figure(figsize=(count + 2, 6))
    else:
        plt.figure(figsize=(n + 2, 6))

    plt.xticks(rotation=90, fontsize=10)
    ax = sns.countplot(
        data=data,
        x=feature,
        palette="Paired",
        order=data[feature].value_counts().index[:n],
    )

    for p in ax.patches:
        if perc == True:
            label = "{:.1f}%".format(
                100 * p.get_height() / total
            )  # percentage of each class of the category
        else:
            label = p.get_height()  # count of each level of the category

        x = p.get_x() + p.get_width() / 2  # width of the plot
        y = p.get_height()  # height of the plot

        ax.annotate(
            label,
            (x, y),
            ha="center",
            va="center",
            size=12,
            xytext=(0, 5),
            textcoords="offset points",
        )  # annotate the percentage

    plt.show()  # show the plot
```

```python
def stacked_barplot(data, predictor, target):
    """
    Print the category counts and plot a stacked bar chart

    data: dataframe
    predictor: independent variable
    target: target variable
    """
    count = data[predictor].nunique()
    sorter = data[target].value_counts().index[-1]
    tab1 = pd.crosstab(data[predictor], data[target], margins=True).sort_values(
        by=sorter, ascending=False
    )
    print(tab1)
    print("-" * 120)
    tab = pd.crosstab(data[predictor], data[target], normalize="index").sort_values(
        by=sorter, ascending=False
    )
    tab.plot(kind="bar", stacked=True, figsize=(count + 5, 5))
    plt.legend(
        loc="lower left", frameon=False,
    )
    plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
    plt.show()
```

```python
### function to plot distributions wrt target


def distribution_plot_wrt_target(data, predictor, target):

    fig, axs = plt.subplots(2, 2, figsize=(10, 7))

    target_uniq = data[target].unique()

    axs[0, 0].set_title("Distribution of target for target=" + str(target_uniq[0]))
    sns.histplot(
        data=data[data[target] == target_uniq[0]],
        x=predictor,
        kde=True,
        ax=axs[0, 0],
        color="teal",
        stat="density",
    )

    axs[0, 1].set_title("Distribution of target for target=" + str(target_uniq[1]))
    sns.histplot(
        data=data[data[target] == target_uniq[1]],
        x=predictor,
        kde=True,
        ax=axs[0, 1],
        color="orange",
        stat="density",
    )

    axs[1, 0].set_title("Boxplot w.r.t target")
    sns.boxplot(data=data, x=target, y=predictor, ax=axs[1, 0], palette="gist_rainbow")

    axs[1, 1].set_title("Boxplot (without outliers) w.r.t target")
    sns.boxplot(
        data=data,
        x=target,
        y=predictor,
        ax=axs[1, 1],
        showfliers=False,
        palette="gist_rainbow",
    )

    plt.tight_layout()
    plt.show()
```

```python
# Test multiple feature combinations to find highest risk
def find_highest_risk_combinations(df, max_combinations=20):
    """Find the highest risk customer profiles"""

    # Define features to test
    feature_combinations = [
        ['Contract', 'InternetService'],
        ['Contract', 'InternetService', 'TechSupport'],
        ['Contract', 'InternetService', 'TechSupport', 'OnlineSecurity'],
        ['Contract', 'InternetService', 'PaymentMethod_New'],
        ['Contract', 'InternetService', 'TechSupport', 'PaymentMethod_New'],
        ['Contract', 'PaymentMethod_New', 'TechSupport', 'OnlineSecurity'],
        ['SeniorCitizen', 'Contract', 'InternetService', 'TechSupport'],
        ['Contract', 'InternetService', 'TechSupport', 'OnlineSecurity', 'OnlineBackup'],
    ]

    all_results = []

    for features in feature_combinations:
        grouped = df.groupby(features).agg({
            'Churn_New': lambda x: (x == 'Yes').sum() / len(x) * 100,
            'gender': 'count'
        }).rename(columns={'gender': 'Customer_Count', 'Churn_New': 'Churn_Rate_%'})

        # Only include combinations with at least 10 customers
        grouped = grouped[grouped['Customer_Count'] >= 10]

        # Get top result
        if len(grouped) > 0:
            top_combo = grouped.sort_values('Churn_Rate_%', ascending=False).head(1)

            # Get the feature values
            combo_values = ' + '.join([f"{feat}={val}" for feat, val in zip(features, top_combo.index[0])])

            all_results.append({
                'Feature_Combination': ' + '.join(features),
                'Profile': combo_values,
                'Churn_Rate_%': top_combo['Churn_Rate_%'].values[0],
                'Customer_Count': int(top_combo['Customer_Count'].values[0])
            })

    results_df = pd.DataFrame(all_results).sort_values('Churn_Rate_%', ascending=False)
    return results_df


print("="*70)
highest_risk = find_highest_risk_combinations(df)
print(highest_risk.head(20))
print("="*70)

# Get THE absolute highest risk profile with sufficient customers
print("\n HIGHEST RISK PROFILE:")
print("-"*70)
top_risk = highest_risk.iloc[0]
print(f"Profile: {top_risk['Profile']}")
print(f"Churn Rate: {top_risk['Churn_Rate_%']:.2f}%")
print(f"Customer Count: {top_risk['Customer_Count']}")
print("="*70)
```

Model evaluation criterion
¶
A Customer Churn prediction model can make incorrect predictions in the following ways:




False Positive :
 The model predicts non-churner as a churner. This may lead to unnecessary retention efforts or offers being extended to customers who would not have churned.


False Negative :
 The model predicts a churner as a non-churner. In this case a customer who is likely to leave is not identified in right time that could result a potential revenue loss.




Which Loss is greater?
¶


False negatives
 are more costly for the company because they represent customers who actually churn but were not flagged by the model.




How to reduce this loss i.e need to reduce False Negatives?
¶


Company wants 
recall
 to be maximized i.e. we need to reduce the number of false negatives.




First, let's create functions to calculate different metrics and confusion matrix so that we don't have to use the same code repeatedly for each model.
¶


The model_performance_classification_statsmodels function will be used to check the model performance of models.


The confusion_matrix_statsmodels function will be used to plot confusion matrix.

```python
# defining a function to compute different metrics to check performance of a classification model built using statsmodels
def model_performance_classification_statsmodels(
    model, predictors, target, threshold=0.5
):
    """
    Function to compute different metrics to check classification model performance

    model: classifier
    predictors: independent variables
    target: dependent variable
    threshold: threshold for classifying the observation as class 1
    """

    # checking which probabilities are greater than threshold
    pred_temp = model.predict(predictors) > threshold
    # rounding off the above values to get classes
    pred = np.round(pred_temp)

    acc = accuracy_score(target, pred)  # to compute Accuracy
    recall = recall_score(target, pred)  # to compute Recall
    precision = precision_score(target, pred)  # to compute Precision
    f1 = f1_score(target, pred)  # to compute F1-score

    # creating a dataframe of metrics
    df_perf = pd.DataFrame(
        {"Accuracy": acc, "Recall": recall, "Precision": precision, "F1": f1,},
        index=[0],
    )

    return df_perf
```

```python
# defining a function to plot the confusion_matrix of a classification model


def confusion_matrix_statsmodels(model, predictors, target, threshold=0.5):
    """
    To plot the confusion_matrix with percentages

    model: classifier
    predictors: independent variables
    target: dependent variable
    threshold: threshold for classifying the observation as class 1
    """
    y_pred = model.predict(predictors) > threshold
    cm = confusion_matrix(target, y_pred)
    labels = np.asarray(
        [
            ["{0:0.0f}".format(item) + "\n{0:.2%}".format(item / cm.flatten().sum())]
            for item in cm.flatten()
        ]
    ).reshape(2, 2)

    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=labels, fmt="")
    plt.ylabel("True label")
    plt.xlabel("Predicted label")
```

There are different ways of detecting (or testing) multi-collinearity, one such way is the Variation Inflation Factor.

Additional Information on VIF
¶


Variance Inflation factor:
 Variance inflation factors measure the inflation in the variances of the regression coefficients estimates due to collinearity that exist among the predictors. It is a measure of how much the variance of the estimated regression coefficient βk is "inflated" by the existence of correlation among the predictor variables in the model.




General Rule of thumb: If VIF is 1 then there is no correlation among the kth predictor and the remaining predictor variables, and hence the variance of β̂k is not inflated at all. Whereas if VIF exceeds 5, we say there is moderate VIF and if it is 10 or exceeding 10, it shows signs of high multi-collinearity. But the purpose of the analysis should dictate which threshold to use.

```python
vif_series = pd.Series(
    [variance_inflation_factor(X_train_sm.values, i) for i in range(X_train_sm.shape[1])],
    index=X_train_sm.columns,
    dtype=float,
)
print("Series before feature selection: \n\n{}\n".format(vif_series))
```

```python
# Create X_train with constant added once to be used throughout the feature selection
X_train_features_with_const = sm.add_constant(X_train.astype(float), prepend=True)

# initial list of columns, including 'const'
cols_for_selection = X_train_features_with_const.columns.tolist()

# setting an initial max p-value
max_p_value = 1

while len(cols_for_selection) > 0:
    # defining the train set for the current iteration
    X_train_aux = X_train_features_with_const[cols_for_selection]

    # fitting the model
    model = sm.Logit(y_train, X_train_aux).fit(disp=False)

    # getting the p-values and the maximum p-value
    p_values = model.pvalues
    max_p_value = p_values.max()

    # name of the variable with maximum p-value
    feature_with_p_max = p_values.idxmax()

    # If the max p-value is greater than 0.05, and it's not the constant, remove the feature
    if max_p_value > 0.05:
        if feature_with_p_max == 'const':
            # If 'const' is the most insignificant, we stop removing features
            # as 'const' is almost always retained.
            break
        else:
            cols_for_selection.remove(feature_with_p_max)
    else:
        break

selected_features = cols_for_selection
print(selected_features)
```

Advanced Model Building
¶

```python
# Defining a function to compute different metrics to check performance of a classification model built using sklearn
def model_performance_classification_sklearn(
    model, predictors, target, threshold=0.5
):
    """
    Function to compute different metrics to check classification model performance

    model: classifier
    predictors: independent variables
    target: dependent variable
    threshold: threshold for classifying the observation as class 1
    """

    # Get probabilities for the positive class
    y_pred_proba = model.predict_proba(predictors)[:, 1]

    # Classify based on the threshold
    y_pred = (y_pred_proba > threshold).astype(int)

    acc = accuracy_score(target, y_pred)  # to compute Accuracy
    recall = recall_score(target, y_pred)  # to compute Recall
    precision = precision_score(target, y_pred)  # to compute Precision
    f1 = f1_score(target, y_pred)  # to compute F1-score

    # creating a dataframe of metrics
    df_perf = pd.DataFrame(
        {"Accuracy": acc, "Recall": recall, "Precision": precision, "F1": f1,},
        index=[0],
    )

    return df_perf
```

Actionable Insights and  Recommandations

Insights:




The analysis identifies several critical factors that drive customer churn at AlphaCom:




Contract Type Influence
: Contract type is a primary indicator of churn risk. Customers on month-to-month contracts exhibit significantly higher churn rates compared to those on one-year or two-year contracts.




Service Bundle Impact:
 Lack of specialized services like Online Security, Tech Support, and Online Backup is strongly correlated with increased churn. Conversely, customers who subscribe to these services are more likely to stay.




Internet Service Quality:
 Customers using Fiber Optic internet services show a higher propensity for churn than those using DSL. This often suggests potential issues with service quality, pricing, or perceived value for fiber customers.




Tenure and Loyalty:
 Customer tenure is a significant predictor; the risk of churn is highest during the initial months of service. As tenure increases, customers generally become more loyal.




Financial Factors:
 High Monthly Charges are a major driver of churn, particularly if customers do not perceive high value for the price.




Payment Methods:
 Customers using Electronic Checks as their payment method tend to churn at a higher rate than those using automated payment methods like bank transfers or credit cards.








Actionable Business Recommendations:




To mitigate churn and improve retention, AlphaCom should implement the following strategies:




Incentivize Long-Term Contracts:




Offer discounts or loyalty rewards to encourage month-to-month customers to migrate to one-year or two-year contracts.


Introduce "early-bird" renewal benefits for customers approaching the end of their long-term contracts.




Promote Security and Support Bundles:




Create service bundles that include Online Security and Tech Support at a discounted rate for new customers.


Target high-risk fiber optic customers with free trials of premium tech support or device protection to increase the "stickiness" of their plans.




Targeted Outreach for Fiber Optic Users:




Conduct targeted surveys specifically for Fiber Optic customers to identify pain points regarding service stability or pricing.


Implement a "proactive outreach" program where technical teams verify line quality for fiber users in high-churn areas.




Focus on Early-Stage Retention:




Implement an onboarding program for the first 3–6 months that includes regular check-ins and tailored tutorials on how to maximize their service value.


Offer a "loyalty bonus" (e.g., a one-time bill credit) after the first 6 months of continuous service.




Optimize Payment and Billing:




Encourage the adoption of Automated Payment Methods (Credit Card/Bank Transfer) by offering a small monthly discount (e.g., $2–$5 off).


Streamline the paperless billing interface to ensure customers can easily understand their monthly charges, reducing frustration that leads to churn.




Conclusion




The analysis of AlphaCom’s customer data demonstrates that churn is not random but driven by specific service gaps and contractual structures. While high monthly charges and short-term contracts are the biggest risks, the data shows that 
bundling technical support and security services
 significantly increases customer retention. By shifting from a reactive approach to a proactive, data-driven strategy—focusing on early-tenure engagement and incentivizing long-term commitments—AlphaCom can safeguard its revenue and significantly reduce the high cost of customer acquisition.
